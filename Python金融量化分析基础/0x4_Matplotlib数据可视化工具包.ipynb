### 2. plot函数周边

```python
# 一个图里包含多条线
# 只需要写多个plot即可
plt.plot([1,2,3,4], [2,3,1,7], "o-r")
plt.plot([1,2,3,4], [3,4,2,1], "^:b")
plt.show()
```

```python
# 一个图里包含多条线
# 只需要写多个plot即可
plt.plot([1,2,3,4], [2,3,1,7], "o-r")
plt.plot([1,2,3,4], [3,4,2,1], "^:b")
plt.title('Matplotlib Test Plot')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.xlim(0, 5)
plt.ylim(0, 8)
plt.show()
```

```python
# 一个图里包含多条线
# 只需要写多个plot即可
plt.plot([1,2,3,4], [2,3,1,7], "o-r")
plt.plot([1,2,3,4], [3,4,2,1], "^:b")
plt.title('Matplotlib Test Plot')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.xlim(0, 5)
plt.ylim(0, 8)
plt.xticks([0, 2, 4])
plt.show()
```

```python
import numpy as np
# 一个图里包含多条线
# 只需要写多个plot即可
plt.plot([1,2,3,4], [2,3,1,7], "o-r", label = "line A")
plt.plot([1,2,3,4], [3,4,2,1], "^:b", label = "line B")
plt.title('Matplotlib Test Plot')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.xlim(0, 5)
plt.ylim(0, 8)

# 图例 上面需要添加 label属性
plt.legend()
plt.xticks(np.arange(0, 10, 2))
plt.show()
```

### 小作业1：


```python
x = np.linspace(-1000, 1000, 100000)
y1 = x.copy()
y2 = x ** 2
y3 = x**3 + 5*x**2 + 2*x + 1
plt.plot(x, y1, color='red', label='y = x')
plt.plot(x, y2, color='g', label='y = x^2')
plt.plot(x, y3, color='b', label='y = x^3 + 5x^2 + 2x + 1')
plt.xlim(-1000, 1000)
plt.ylim(-1000, 1000)
plt.legend()
plt.show()
```

### 4. 画布与子图


```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # 将画布分为2 * 2, ax1在第一个位置
ax1.plot([1,2,3,4], [5,6,7,8])
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax1.plot([1,2,3,4], [5,6,7,8])
ax2 = fig.add_subplot(2,1,2)
plt.show()
```

### 5. 柱状图、饼图
Matplotlib支持的图类型



```python
data = [32, 48, 21, 100]
labels = ['Jan', 'Feb', 'Mar', 'Apr']
plt.bar(np.arange(len(data)), data, color=['r','g','b', 'k'] ,width = 0.5)
plt.xticks(np.arange(len(data)), labels)
plt.show()
```

```python
plt.pie([10, 20, 30, 40], labels = ['a', 'b', 'c', 'd'], autopct="%.2f%%")
plt.show()
```

```python
plt.pie([10, 20, 30, 40], labels = ['a', 'b', 'c', 'd'], autopct="%.2f%%", explode=[0.2,0,0.1, 0])
plt.axis('equal')
plt.show()
```

### 6. 绘制K线图

输入`from matplotlib.finance import candlestick_ohlc`

发现有报错，仔细勘查后发现有两个问题，一个是matplotlib模块已经剔除了，所以得额外再安装

解决方式

`pip install https://github.com/matplotlib/mpl_finance/archive/master.zip`
使用`import mpl_finance as fin`来调用

```python
import mpl_finance as fin
from matplotlib.dates import date2num # 用于将date转换为float类型的时间戳
```

```python
df
```

```python
df['time'] = date2num(df.index.to_pydatetime())
df
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
arr = df[['time', 'open', 'close', 'high', 'low']].values
fin.candlestick_ochl(ax, arr) # ochl -> open close high low
fig.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
arr = df['2017-1'][['time', 'open', 'close', 'high', 'low']].values
fin.candlestick_ochl(ax, arr) # ochl -> open close high low
fig.show()
```
