### 5 布尔型索引

问题：给一个数组，选出数组中所有大于5的数

```python
import random
a = [random.randint(0, 10) for i in range(20)]
a
```

```python
# 传统方式
list(filter(lambda x: x > 5, a))
```

```python
a = np.array(a)
a
```

```python
# 布尔型索引
a[a > 5]
```

```python
a > 5
```

```python
# 返回大于5的偶数
a[(a > 5) & (a%2==0)]
```
