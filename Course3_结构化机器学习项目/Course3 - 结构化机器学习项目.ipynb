## 2.7 迁移学习 Transfer learning
* Pre-train  预训练
* fine tuning 微调

预训练先快速部署一个模型（已经训练过的），微调就是在这个基础上做一些修改，用新数据去训练修改的这几层，达到快速应用的目的，不用从头训练一个新模型

> 迁移学习什么时候是有意义的呢？
* 迁移学习起作用的场合是**迁移来源问题**你有很多数据，但**迁移目标问题**没有那么多数据。

## 2.8 多任务学习 Multi-task learning
* Multi-task learning 并行
* Transfer learning 串行
---
> 多任务学习什么时候是有意义的呢？
* 如果你训练的一组任务可以共有低层次特征
* Usually: 如果每个任务的数据量很接近
* 可以训练一个足够大的神经网络同时做好所有的工作

相比于训练单个神经网络，多任务学习会降低性能的唯一情况是你的神经网络还不够大。但如果你训练一个足够大的神经网络，那么多任务学习，几乎不会降低性能。一般情况比单独训练神经网络来单独完成各个任务性能要好。

其中一个应用：计算机视觉 物体检测 人们经常训练一个神经网络同时检测很多不同的物体，这比训练单独的神经网络来检测物体要更好。

## 2.10 是否要使用端到端的深度学习
Pros and cons of end-to-end deep learning:
> Pros
* Let the data speak (如果你有足够的X,Y数据，那么不管从x到y最合适的函数映射是什么，你训练一个足够大的神经网络并期望神经网路自己解决)
* Less hand-designing of components needed (简化你的设计工作流程)

> Cons
* May need large amount of data
* Excludes potentially useful hand-designed components 排除了可能有用的手工设计组件
